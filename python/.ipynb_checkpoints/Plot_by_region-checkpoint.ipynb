```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=data = pd.read_csv('Mais_D1.csv')
df
```

```python
var='chirps_Prec'
```

```python
plt.figure(figsize=(15,10))

Mouhoun=df.loc[df['province'].isin(['Bale','Banwa','Nayala','Mouhoun'])]
Mouhoun=Mouhoun.groupby(['year']).mean()
y=Mouhoun.index
m1=Mouhoun[var]
plt.plot(y, m1,label='Mouhoun')


Cascade=df.loc[df['province'].isin(['Comoe','Leraba'])]
Cascade=Cascade.groupby(['year']).mean()
m2=Cascade[var]
plt.plot(y, m2,label='Cascade')

C_est=df.loc[df['province'].isin(['Boulgou','Koulpelogo','Kouritenga'])]
C_est=C_est.groupby(['year']).mean()
m3=C_est[var]
plt.plot(y, m3,label='C_est')

Haut_Bassin=df.loc[df['province'].isin(['Tuy','Kenedougou','Houet'])]
Haut_Bassin=Haut_Bassin.groupby(['year']).mean()
m12=Haut_Bassin[var]
plt.plot(y, m12,label='Haut_Bassin')

C_Nord=df.loc[df['province'].isin(['Bam','Namentenga','Sanmatenga'])]
C_Nord=C_Nord.groupby(['year']).mean()
m4=C_Nord[var]
plt.plot(y, m4,label='C_Nord')

C_ouest=df.loc[df['province'].isin(['Boulkiemde','Sanguie','Sissili','Ziro'])]
C_ouest=C_ouest.groupby(['year']).mean()
m5=C_ouest[var]
plt.plot(y, m5,label='C_ouest')

C_Sud=df.loc[df['province'].isin(['Bazega','Nahouri','Zoundweogo'])] 
C_Sud=C_Sud.groupby(['year']).mean()
m6=C_Sud[var]
plt.plot(y, m6,label='C_Sud')

Centre=df.loc[df['province'].isin(['Kadiogo'])] 
Centre=Centre.groupby(['year']).mean()
m7=Centre[var]
plt.plot(y, m7,label='Centre')

Plateau_central=df.loc[df['province'].isin(['Ganzourgou','Kourweogo','Oubritenga'])]
Plateau_central=Plateau_central.groupby(['year']).mean()
m8=Plateau_central[var]
plt.plot(y, m8,label='Plateau_central')

Sahel=df.loc[df['province'].isin(['Oudalan','Seno','Soum','Yagha'])]
Sahel=Sahel.groupby(['year']).mean()
m9=Sahel[var]
plt.plot(y, m9,label='Sahel')

Nord=df.loc[df['province'].isin(['Loroum','Passore','Yatenga','Zondoma'])]
Nord=Nord.groupby(['year']).mean()
m10=Nord[var]
plt.plot(y, m10,label='Nord')

Sud_ouest=df.loc[df['province'].isin(['Poni','Noumbiel','Ioba','Bougouriba'])]
Sud_ouest=Sud_ouest.groupby(['year']).mean()
m11=Sud_ouest[var]
plt.plot(y, m11,label='Sud_ouest')



Est=df.loc[df['province'].isin(['Gnagna','Gourma','Komandjoari','Kompienga','Tapoa'])]
Est=Est.groupby(['year']).mean()
m13=Est[var]
plt.plot(y, m13,label='Est')

plt.legend()
plt.show()
```
