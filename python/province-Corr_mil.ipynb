```python
def model(df,n):

    catvar=['province']
    dummyvars = pd.get_dummies(df[catvar])
    dt= df.iloc[:, 1:]
    
    dt = pd.DataFrame(StandardScaler().fit_transform(dt))
    dt=sc=dt.rename(columns={0:'year', 1:'driver.TSA', 2:'Driver.TNA', 3:'sahel_rainfall', 4:'QBO', 5:'NAO',
       6:'gmsst', 7:'EA', 8:'AMM_sst', 9:'Driver.NTA', 10:'AMO_sst', 11:'Driver.AO', 12:'PDO',
       13:'PRCPTOT', 14:'R10mm', 15:'R1mm', 16:'R20mm', 17:'R5mm', 18:'R95pTOT', 19:'R99pTOT',
       20:'RX1day', 21:'RX5day',22: 'SDII', 23:'CWD', 24:'CDD', 25:'TNn', 26:'TNx', 27:'TXx', 28:'TXn',
       29:'Chirts_Tmax_Temp', 30:'Chirts_Tmin_Temp', 31:'ERA5_Pressure', 32:'ERA5_wins',
       33:'chirps_Prec', 34:'ERA5_soil_moisture', 35:'NDVI', 36:'Mil.rendement'})
    
    
    cor=pd.DataFrame(pd.DataFrame(dt.corr()).iloc[:,36])
    
    absolute=cor['Mil.rendement'].abs()

    sort=pd.DataFrame(absolute).sort_values(by=['Mil.rendement'],ascending=False)
    bbb=sort.loc[sort['Mil.rendement'] >= .2]
    index_list = bbb.index.tolist()
    data=sc[index_list]
    
    data_end = pd.concat([data,dummyvars], axis = 1)
    
    nb= len(data_end.columns)
    print(nb)
    X=data_end.iloc[:,1:]
    y=data_end.iloc[:,0]

    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=n) 

    model1 = RidgeCV()
    model2 = Lasso()
    model3 = RandomForestRegressor(n_estimators = 2000,max_depth=82)
    model4 = DecisionTreeRegressor(max_depth=80) 

    model1.fit(X_train, y_train)
    model2.fit(X_train, y_train)
    model3.fit(X_train, y_train)
    model4.fit(X_train, y_train)

    train_RidgeCV=model1.score(X_train, y_train)
    test_RidgeCV=model1.score(X_test, y_test)

    train_LassoCV=model2.score(X_train, y_train)
    test_LassoCV=model2.score(X_test, y_test)

    train_RandomForest=model3.score(X_train, y_train)
    test_RandomForest=model3.score(X_test, y_test)

    train_DecisionTree=model4.score(X_train, y_train)
    test_DecisionTree=model4.score(X_test, y_test)


    data1 = {'model': ['RandomForest', 'DecisionTree', 'RidgeCV', 'Lasso'],
        'Score_train': [train_RandomForest, train_DecisionTree, train_RidgeCV, train_LassoCV],
        'score_test': [test_RandomForest, test_DecisionTree, test_RidgeCV, test_LassoCV]
        }

   # data = {'model': ['RandomForest', 'DecisionTree'],
     #   'Score_train': [train_RandomForest, train_DecisionTree],
     #   'score_test': [test_RandomForest, test_DecisionTree]
     #   }

    d = pd.DataFrame(data1)
    return d
```

```python
model(df,n=3)
```

```python
def model_province(dt,n,level):
    m=level
    m=np.repeat(m,4, axis=0)
    m={'province':m}
    m=pd.DataFrame(m)
    m=pd.concat([m,model(dt,n,level)],axis = 1)
    return m
```

```python
model_province(df,3,'Ganzourgou')
```

```python
def Data(dt,n,v):
    D=pd.DataFrame()
    for i in range(0,len(v)):
        P=v[i]
        D=pd.concat([D,model_province(dt,n,P)],axis = 0)
    return D
```
